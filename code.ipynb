```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute  import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
```

```python
data = pd.read_csv('possum.csv')
data.head(10)
```

```python
data['sex'].unique()
```

```python
encoder = LabelEncoder()

#Vic = 0 , Other = 1 :
data['Pop'] = encoder.fit_transform(data['Pop'])

# male = 0 , female = 1
data['sex'] = encoder.fit_transform(data['sex'])

data.head(10)
```

```python
feature_not_zero = ['age' , 'footlgth']
imputer = SimpleImputer(strategy = 'mean')
data[feature_not_zero] = imputer.fit_transform(data[feature_not_zero])
```

```python
data.head(50)
```

```python
x = data.iloc[: , :-1]
y = data.iloc[: , -1]
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
norma = StandardScaler()
x_train = norma.fit_transform(x_train)
x_test = norma.transform(x_test)
```

```python
lr = LinearRegression()

lr.fit(x_train , y_train)

pred = lr.predict(x_test)
score = lr.score(x_test , y_test)
score
```

```python
from sklearn.linear_model import SGDRegressor
clf = SGDClassifier(loss="squared_error", penalty="l2", max_iter=5)
clf.fit(x_train, y_train)
SGDClassifier(max_iter=5)
```

```python
knnr = KNeighborsRegressor(n_neighbors=3)

knnr.fit(x_train , y_train)

knnr.score(x_train , y_train)
```
